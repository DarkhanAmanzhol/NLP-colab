```python
from flask_ngrok import run_with_ngrok
from __future__ import unicode_literals
from flask import Flask,render_template,url_for,request
#from nltk_summarization import text_summarizer
from summarization import nltk_summarizer
from grammar_check import grammar_check, similar
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
app = Flask(__name__)

run_with_ngrok(app)
@app.route('/')
def index():
    return render_template('index.html')

@app.route('/analyze',methods=['GET','POST'])
def analyze():
    if request.method == 'POST':
        rawtext = request.form['rawtext']
        percentage = int(request.form.get('percentage'))
        final_summary, language = nltk_summarizer(rawtext, percentage)
    return render_template('index.html',ctext=rawtext,final_summary=final_summary, text_language=language)


@app.route('/compare_summary')
def compare_summary():
    return render_template('compare_summary.html')

@app.route('/comparer',methods=['GET','POST'])
def comparer():
    if request.method == 'POST':
        rawtext = request.form['rawtext']
        corrected_text = grammar_check(rawtext)
        difference = similar(rawtext, corrected_text)

        return render_template('compare_summary.html', ctext=rawtext, final_text=corrected_text, diff=difference)

app.run()
```
